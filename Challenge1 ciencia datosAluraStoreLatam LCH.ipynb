# 3. Calificación promedio de la tienda


```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs de los archivos CSV
urls = {
    "Tienda 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Diccionario para almacenar los promedios por tienda
promedios_calificacion = {}

# Procesar cada tienda
for nombre_tienda, url in urls.items():
    try:
        # Leer el archivo CSV
        df = pd.read_csv(url)

        # Calcular el promedio de calificación
        promedio = df['Calificación'].mean()

        # Almacenar el resultado
        promedios_calificacion[nombre_tienda] = promedio

        print(f"Calificación promedio {nombre_tienda}: {promedio:.2f}")
    except Exception as e:
        print(f"\nError al procesar {nombre_tienda}: {e}")
        promedios_calificacion[nombre_tienda] = None  # Para mantener consistencia

# Filtrar tiendas con datos válidos
tiendas_validas = {k: v for k, v in promedios_calificacion.items() if v is not None}

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(tiendas_validas.keys(), tiendas_validas.values(), color=['#4CAF50', '#2196F3', '#FFC107', '#FF5722'])

# Añadir etiquetas y título
plt.title('Calificación Promedio por Tienda', fontsize=16, pad=20)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)
plt.ylim(0, 5)  # Asumiendo que la calificación va de 0 a 5

# Añadir los valores en las barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2f}',
             ha='center', va='bottom', fontsize=11)

# Añadir línea de referencia para el promedio general si hay datos
if tiendas_validas:
    promedio_general = sum(tiendas_validas.values())/len(tiendas_validas)
    plt.axhline(y=promedio_general, color='red', linestyle='--', linewidth=1)
    plt.text(len(tiendas_validas)-0.5, promedio_general+0.1,
             f'Promedio General: {promedio_general:.2f}',
             color='red', ha='right')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# URLs de los archivos CSV
urls = {
    "Tienda 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Diccionarios para almacenar resultados
mas_vendidos = {}
menos_vendidos = {}

# Procesar cada tienda
for nombre_tienda, url in urls.items():
    try:
        # Leer datos
        df = pd.read_csv(url)

        # Contar ventas por producto
        ventas_por_producto = df['Producto'].value_counts()

        # Obtener producto más vendido
        top_producto = ventas_por_producto.idxmax()
        top_cantidad = ventas_por_producto.max()
        mas_vendidos[nombre_tienda] = (top_producto, top_cantidad)

        # Obtener producto menos vendido (excluyendo ceros si existen)
        ventas_positivas = ventas_por_producto[ventas_por_producto > 0]
        if not ventas_positivas.empty:
            bottom_producto = ventas_positivas.idxmin()
            bottom_cantidad = ventas_positivas.min()
            menos_vendidos[nombre_tienda] = (bottom_producto, bottom_cantidad)
        else:
            menos_vendidos[nombre_tienda] = ("Ninguno", 0)

        print(f"\n{nombre_tienda}:")
        print(f"  Más vendido: {top_producto} ({top_cantidad} unidades)")
        print(f"  Menos vendido: {menos_vendidos[nombre_tienda][0]} ({menos_vendidos[nombre_tienda][1]} unidades)")

    except Exception as e:
        print(f"\nError procesando {nombre_tienda}: {str(e)}")

# Preparar datos para el gráfico
tiendas = list(mas_vendidos.keys())
productos_top = [x[0] for x in mas_vendidos.values()]
cantidades_top = [x[1] for x in mas_vendidos.values()]
productos_bottom = [x[0] for x in menos_vendidos.values()]
cantidades_bottom = [x[1] for x in menos_vendidos.values()]

# Configurar el gráfico
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
fig.suptitle('Productos Más y Menos Vendidos por Tienda', fontsize=16)

# Gráfico de productos más vendidos
ax1.bar(tiendas, cantidades_top, color='#2ecc71')
ax1.set_title('Productos Más Vendidos', fontsize=14)
ax1.set_ylabel('Cantidad Vendida', fontsize=12)
for i, (prod, cant) in enumerate(zip(productos_top, cantidades_top)):
    ax1.text(i, cant+0.5, f"{prod}\n({cant} u)", ha='center', va='bottom')

# Gráfico de productos menos vendidos
ax2.bar(tiendas, cantidades_bottom, color='#e74c3c')
ax2.set_title('Productos Menos Vendidos', fontsize=14)
ax2.set_ylabel('Cantidad Vendida', fontsize=12)
for i, (prod, cant) in enumerate(zip(productos_bottom, cantidades_bottom)):
    ax2.text(i, cant+0.1, f"{prod}\n({cant} u)", ha='center', va='bottom')

# Ajustar layout
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
# Función para añadir un mapa adicional que muestre todas las tiendas simultáneamente
def crear_mapa_comparativo(datos):
    # Verificar si hay datos con coordenadas válidas
    datos_validos = datos.dropna(subset=['lat', 'lon'])

    if datos_validos.empty:
        print("No hay datos con coordenadas válidas para crear el mapa.")
        return None

    # Calcular centro del mapa
    centro_lat = datos_validos['lat'].mean()
    centro_lon = datos_validos['lon'].mean()

    # Crear mapa base
    mapa = folium.Map(location=[centro_lat, centro_lon], zoom_start=4)

    # Definir colores para cada tienda
    colores_tiendas = {
        'Tienda 1': '#FF5733',  # Rojo-naranja
        'Tienda 2': '#33FF57',  # Verde
        'Tienda 3': '#3357FF',  # Azul
        'Tienda 4': '#F033FF'   # Morado
    }

    # Para cada tienda, agregar marcadores agrupados
    for tienda_id in datos_validos['tienda_id'].unique():
        # Filtrar datos de esta tienda
        datos_tienda = datos_validos[datos_validos['tienda_id'] == tienda_id]

        # Crear un cluster para esta tienda
        cluster = MarkerCluster(name=tienda_id).add_to(mapa)

        # Agrupar puntos cercanos para esta tienda
        datos_tienda['ubicacion'] = datos_tienda.apply(lambda row: f"{row['lat']:.3f}_{row['lon']:.3f}", axis=1)
        ubicaciones = datos_tienda.groupby('ubicacion').agg(
            lat=('lat', 'first'),
            lon=('lon', 'first'),
            total_ventas=('Producto', 'count'),
            ingreso_total=('ingreso_total', 'sum')
        ).reset_index()

        # Color para esta tiendaimport pandas as pd
import folium
import requests
from io import StringIO
import numpy as np
from folium.plugins import MarkerCluster
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from IPython.display import display, HTML

# URLs de los archivos CSV
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
]

# Función para descargar y limpiar los datos
def obtener_datos(urls):
    dataframes = []

    for i, url in enumerate(urls):
        try:
            # Descargar datos
            response = requests.get(url)
            response.raise_for_status()

            # Leer los datos en un DataFrame
            df = pd.read_csv(StringIO(response.text), sep=',')

            # Añadir identificador de tienda
            df['tienda_id'] = f'Tienda {i+1}'

            # Convertir columnas numéricas
            for col in ['Precio', 'Costo de envío']:
                df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '.'), errors='coerce')

            # Limpiar columnas de latitud y longitud
            for col in ['lat', 'lon']:
                # Verificar si la columna existe
                if col in df.columns:
                    df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '.'), errors='coerce')

            # Calcular el ingreso total por venta
            df['ingreso_total'] = df['Precio'] + df['Costo de envío']

            dataframes.append(df)

        except Exception as e:
            print(f"Error al procesar la tienda {i+1}: {str(e)}")

    # Concatenar todos los dataframes
    if dataframes:
        return pd.concat(dataframes, ignore_index=True)
    else:
        return None

# Función para crear mapa de ingresos con comparación entre tiendas
def crear_mapa_ingresos(datos):
    # Verificar si hay datos con coordenadas válidas
    datos_validos = datos.dropna(subset=['lat', 'lon'])

    if datos_validos.empty:
        print("No hay datos con coordenadas válidas para crear el mapa.")
        return None

    # Calcular el centro del mapa
    centro_lat = datos_validos['lat'].mean()
    centro_lon = datos_validos['lon'].mean()

    # Crear el mapa
    mapa = folium.Map(location=[centro_lat, centro_lon], zoom_start=4)

    # Definir colores para cada tienda
    colores_tiendas = {
        'Tienda 1': '#FF5733',  # Rojo-naranja
        'Tienda 2': '#33FF57',  # Verde
        'Tienda 3': '#3357FF',  # Azul
        'Tienda 4': '#F033FF'   # Morado
    }

    # Crear grupos de marcadores para cada tienda
    grupos_tiendas = {}
    for tienda_id in datos_validos['tienda_id'].unique():
        grupo = folium.FeatureGroup(name=tienda_id)
        grupos_tiendas[tienda_id] = grupo
        mapa.add_child(grupo)

    # Obtener el rango de ingresos para determinar el tamaño de los marcadores
    min_ingreso = datos_validos['ingreso_total'].min()
    max_ingreso = datos_validos['ingreso_total'].max()

    # Agregar marcadores para cada venta
    for idx, row in datos_validos.iterrows():
        # Determinar el color según la tienda
        color = colores_tiendas.get(row['tienda_id'], '#000000')

        # Normalizar el ingreso para determinar el tamaño
        norm_ingreso = (row['ingreso_total'] - min_ingreso) / (max_ingreso - min_ingreso) if max_ingreso > min_ingreso else 0.5
        radio = 5 + (15 * norm_ingreso)

        # Crear popup con información de la venta
        popup_text = f"""
        <b>Tienda:</b> {row['tienda_id']}<br>
        <b>Producto:</b> {row['Producto']}<br>
        <b>Categoría:</b> {row['Categoría del Producto']}<br>
        <b>Precio:</b> {row['Precio']}<br>
        <b>Costo de envío:</b> {row['Costo de envío']}<br>
        <b>Ingreso total:</b> {row['ingreso_total']}<br>
        <b>Fecha:</b> {row['Fecha de Compra']}<br>
        <b>Vendedor:</b> {row['Vendedor']}<br>
        <b>Lugar:</b> {row['Lugar de Compra']}<br>
        """

        # Crear marcador y añadirlo al grupo correspondiente
        folium.CircleMarker(
            location=[row['lat'], row['lon']],
            radius=radio,
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.7,
            popup=folium.Popup(popup_text, max_width=300)
        ).add_to(grupos_tiendas[row['tienda_id']])

    # Agregar control de capas para mostrar/ocultar tiendas
    folium.LayerControl().add_to(mapa)

    # Crear leyenda personalizada para tiendas e ingresos
    legend_html = '''
    <div style="position: fixed;
                bottom: 50px; left: 50px; width: 230px;
                border:2px solid grey; z-index:9999; font-size:14px;
                background-color:white; padding: 10px;
                border-radius: 5px;">
    <p style="margin: 0"><b>Tiendas</b></p>
    '''

    # Añadir colores de tiendas a la leyenda
    for tienda_id, color in colores_tiendas.items():
        legend_html += f'''
        <div style="display: flex; align-items: center; margin-top: 5px;">
            <div style="width: 20px; height: 20px; background-color:{color}; border-radius: 50%; margin-right: 5px;"></div>
            <span>{tienda_id}</span>
        </div>
        '''

    # Añadir información de tamaño según ingreso
    legend_html += '''
    <p style="margin-top: 15px; margin-bottom: 5px;"><b>Tamaño = Ingreso Total</b></p>
    <div style="display: flex; align-items: center; margin-top: 5px;">
        <div style="width: 10px; height: 10px; border-radius: 50%; border: 1px solid #444; margin-right: 5px;"></div>
        <span>Bajo</span>
    </div>
    <div style="display: flex; align-items: center; margin-top: 5px;">
        <div style="width: 15px; height: 15px; border-radius: 50%; border: 1px solid #444; margin-right: 5px;"></div>
        <span>Medio</span>
    </div>
    <div style="display: flex; align-items: center; margin-top: 5px;">
        <div style="width: 20px; height: 20px; border-radius: 50%; border: 1px solid #444; margin-right: 5px;"></div>
        <span>Alto</span>
    </div>
    </div>
    '''

    mapa.get_root().html.add_child(folium.Element(legend_html))

    return mapa

# Función para crear mapa de cantidad de ventas con comparación entre tiendas
def crear_mapa_cantidad_ventas(datos):
    # Verificar si hay datos con coordenadas válidas
    datos_validos = datos.dropna(subset=['lat', 'lon'])

    if datos_validos.empty:
        print("No hay datos con coordenadas válidas para crear el mapa.")
        return None

    # Calcular el centro del mapa
    centro_lat = datos_validos['lat'].mean()
    centro_lon = datos_validos['lon'].mean()

    # Crear el mapa
    mapa = folium.Map(location=[centro_lat, centro_lon], zoom_start=4)

    # Definir colores para cada tienda
    colores_tiendas = {
        'Tienda 1': '#FF5733',  # Rojo-naranja
        'Tienda 2': '#33FF57',  # Verde
        'Tienda 3': '#3357FF',  # Azul
        'Tienda 4': '#F033FF'   # Morado
    }

    # Agrupar datos por ubicación y tienda para contar ventas
    datos_validos['ubicacion'] = datos_validos.apply(lambda row: f"{row['lat']:.5f}_{row['lon']:.5f}", axis=1)

    # Crear grupos de capas para cada tienda
    grupos_tiendas = {}
    for tienda_id in datos_validos['tienda_id'].unique():
        grupo = folium.FeatureGroup(name=tienda_id)
        grupos_tiendas[tienda_id] = grupo
        mapa.add_child(grupo)

    # Para cada tienda, procesar sus datos separadamente
    for tienda_id in datos_validos['tienda_id'].unique():
        # Filtrar datos por tienda
        datos_tienda = datos_validos[datos_validos['tienda_id'] == tienda_id]

        # Contar ventas por ubicación
        conteo_tienda = datos_tienda.groupby('ubicacion').size().reset_index(name='cantidad_ventas')

        # Unir el conteo con los datos originales para obtener coordenadas
        datos_agrupados = conteo_tienda.merge(
            datos_tienda[['ubicacion', 'lat', 'lon']].drop_duplicates('ubicacion'),
            on='ubicacion'
        )

        # Obtener el rango de cantidad de ventas para esta tienda
        min_ventas = conteo_tienda['cantidad_ventas'].min()
        max_ventas = conteo_tienda['cantidad_ventas'].max()

        # Obtener el color base para esta tienda
        color_base = colores_tiendas.get(tienda_id, '#000000')

        # Añadir marcadores para esta tienda
        for idx, row in datos_agrupados.iterrows():
            # Normalizar la cantidad para determinar la opacidad
            norm_cantidad = (row['cantidad_ventas'] - min_ventas) / (max_ventas - min_ventas) if max_ventas > min_ventas else 0.5

            # Determinar el radio basado en la cantidad
            radio = 10 + (20 * norm_cantidad)

            # Crear popup con información
            popup_text = f"""
            <b>Tienda:</b> {tienda_id}<br>
            <b>Ubicación:</b> {row['lat']}, {row['lon']}<br>
            <b>Cantidad de ventas:</b> {row['cantidad_ventas']}<br>
            """

            folium.CircleMarker(
                location=[row['lat'], row['lon']],
                radius=radio,
                color=color_base,
                fill=True,
                fill_color=color_base,
                fill_opacity=0.7,
                popup=folium.Popup(popup_text, max_width=300)
            ).add_to(grupos_tiendas[tienda_id])

    # Agregar control de capas para mostrar/ocultar tiendas
    folium.LayerControl().add_to(mapa)

    # Crear leyenda personalizada para tiendas y cantidad de ventas
    legend_html = '''
    <div style="position: fixed;
                bottom: 50px; left: 50px; width: 230px;
                border:2px solid grey; z-index:9999; font-size:14px;
                background-color:white; padding: 10px;
                border-radius: 5px;">
    <p style="margin: 0"><b>Tiendas</b></p>
    '''

    # Añadir colores de tiendas a la leyenda
    for tienda_id, color in colores_tiendas.items():
        legend_html += f'''
        <div style="display: flex; align-items: center; margin-top: 5px;">
            <div style="width: 20px; height: 20px; background-color:{color}; border-radius: 50%; margin-right: 5px;"></div>
            <span>{tienda_id}</span>
        </div>
        '''

    # Añadir información de tamaño según cantidad de ventas
    legend_html += '''
    <p style="margin-top: 15px; margin-bottom: 5px;"><b>Tamaño = Cantidad de Ventas</b></p>
    <div style="display: flex; align-items: center; margin-top: 5px;">
        <div style="width: 10px; height: 10px; border-radius: 50%; border: 1px solid #444; margin-right: 5px;"></div>
        <span>Baja</span>
    </div>
    <div style="display: flex; align-items: center; margin-top: 5px;">
        <div style="width: 20px; height: 20px; border-radius: 50%; border: 1px solid #444; margin-right: 5px;"></div>
        <span>Media</span>
    </div>
    <div style="display: flex; align-items: center; margin-top: 5px;">
        <div style="width: 30px; height: 30px; border-radius: 50%; border: 1px solid #444; margin-right: 5px;"></div>
        <span>Alta</span>
    </div>
    </div>
    '''

    mapa.get_root().html.add_child(folium.Element(legend_html))

    return mapa

# Función principal
def main():
    print("Descargando y procesando datos...")
    datos = obtener_datos(urls)

    if datos is None or datos.empty:
        print("No se pudieron obtener datos válidos.")
        return

    print(f"Se procesaron {len(datos)} registros de ventas.")

    # Mostrar un resumen por tienda
    print("\nResumen por tienda:")
    resumen_tiendas = datos.groupby('tienda_id').agg(
        total_ventas=('Producto', 'count'),
        ingreso_promedio=('ingreso_total', 'mean'),
        ingreso_total=('ingreso_total', 'sum')
    ).reset_index()

    for _, row in resumen_tiendas.iterrows():
        print(f"{row['tienda_id']}: {row['total_ventas']} ventas, Ingreso promedio: {row['ingreso_promedio']:.2f}, Ingreso total: {row['ingreso_total']:.2f}")

    # Crear el mapa de ingresos
    print("\nCreando mapa de ingresos por tienda...")
    mapa_ingresos = crear_mapa_ingresos(datos)
    if mapa_ingresos:
        # Mostrar el mapa directamente en el notebook
        display(HTML("<h2>Mapa de ingresos por zona geográfica (colores por tienda)</h2>"))
        display(HTML("<p><b>Instrucciones:</b> Utilice el control de capas (esquina superior derecha) para mostrar/ocultar tiendas específicas.</p>"))
        display(mapa_ingresos)

    # Crear el mapa de cantidad de ventas
    print("Creando mapa de cantidad de ventas por tienda...")
    mapa_cantidad = crear_mapa_cantidad_ventas(datos)
    if mapa_cantidad:
        # Mostrar el mapa directamente en el notebook
        display(HTML("<h2>Mapa de cantidad de ventas por zona geográfica (colores por tienda)</h2>"))
        display(HTML("<p><b>Instrucciones:</b> Utilice el control de capas (esquina superior derecha) para mostrar/ocultar tiendas específicas.</p>"))
        display(mapa_cantidad)

    # Crear un análisis comparativo adicional
    print("\nCreando análisis comparativo de tiendas...")

    # Análisis por categoría de producto
    categorias_por_tienda = datos.groupby(['tienda_id', 'Categoría del Producto']).size().reset_index(name='cantidad')
    top_categorias_tienda = categorias_por_tienda.sort_values(['tienda_id', 'cantidad'], ascending=[True, False]).groupby('tienda_id').head(3)

    print("\nTop 3 categorías de productos por tienda:")
    for tienda_id in top_categorias_tienda['tienda_id'].unique():
        print(f"\n{tienda_id}:")
        categorias = top_categorias_tienda[top_categorias_tienda['tienda_id'] == tienda_id]
        for _, cat_row in categorias.iterrows():
            print(f"  - {cat_row['Categoría del Producto']}: {cat_row['cantidad']} ventas")

    # Las 5 categorías más vendidas en total
    top_categorias = datos['Categoría del Producto'].value_counts().head(5)
    print("\nTop 5 categorías más vendidas en total:")
    for cat, count in top_categorias.items():
        print(f"- {cat}: {count} ventas")

    # También puedes guardar los mapas como HTML si lo deseas
    mapa_ingresos.save('mapa_ingresos_tiendas.html')
    mapa_cantidad.save('mapa_cantidad_ventas_tiendas.html')
    print("\nLos mapas también se han guardado como archivos HTML ('mapa_ingresos_tiendas.html' y 'mapa_cantidad_ventas_tiendas.html')")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

# URLs de los archivos CSV
urls = {
    "Tienda 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Diccionario para almacenar los KPIs
kpis_tiendas = {}

# Procesar cada tienda
for nombre_tienda, url in urls.items():
    try:
        df = pd.read_csv(url)

        # Calcular KPIs básicos
        ingresos_totales = df['Precio'].sum()
        costo_envio_promedio = df['Costo de envío'].mean()
        ingresos_netos = ingresos_totales - df['Costo de envío'].sum()
        calificacion_promedio = df['Calificación'].mean()  # Sin redondear

        # Calcular categorías más y menos vendidas
        categorias_counts = df['Categoría del Producto'].value_counts()
        categoria_mas_vendida = f"{categorias_counts.idxmax()} ({categorias_counts.max()} ventas)"
        categoria_menos_vendida = f"{categorias_counts.idxmin()} ({categorias_counts.min()} ventas)"

        # Calcular productos más y menos vendidos
        productos_counts = df['Producto'].value_counts()
        producto_mas_vendido = f"{productos_counts.idxmax()} ({productos_counts.max()} unidades)"
        producto_menos_vendido = f"{productos_counts.idxmin()} ({productos_counts.min()} unidades)"

        # Almacenar KPIs
        kpis_tiendas[nombre_tienda] = {
            'Ingresos Totales': ingresos_totales,
            'Costo Envío Promedio': costo_envio_promedio,
            'Ingresos Netos': ingresos_netos,
            'Calificación Promedio': calificacion_promedio,  # Valor exacto
            'Categoría Más Vendida': categoria_mas_vendida,
            'Categoría Menos Vendida': categoria_menos_vendida,
            'Producto Más Vendido': producto_mas_vendido,
            'Producto Menos Vendido': producto_menos_vendido
        }

    except Exception as e:
        print(f"Error procesando {nombre_tienda}: {e}")

# Crear DataFrame comparativo
df_comparativo = pd.DataFrame.from_dict(kpis_tiendas, orient='index')

# Función para formatear valores monetarios
def format_currency(x):
    if isinstance(x, (int, float)):
        return f"${x:,.2f}"
    return x

# Aplicar formato solo a las columnas monetarias
df_comparativo[['Ingresos Totales', 'Costo Envío Promedio', 'Ingresos Netos']] = (
    df_comparativo[['Ingresos Totales', 'Costo Envío Promedio', 'Ingresos Netos']]
    .applymap(format_currency)
)

# Mostrar tabla comparativa con estilo
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: lightgreen' if v else '' for v in is_max]

def highlight_min(s):
    is_min = s == s.min()
    return ['background-color: lightcoral' if v else '' for v in is_min]

styled_table = (
    df_comparativo.style
    .set_properties(**{'text-align': 'center', 'border': '1px solid black'})
    .set_table_styles([{
        'selector': 'th',
        'props': [('background-color', '#f7f7f7'), ('font-weight', 'bold')]
    }])
    .apply(highlight_max, subset=['Ingresos Totales', 'Ingresos Netos', 'Calificación Promedio'])
    .apply(highlight_min, subset=['Costo Envío Promedio'])
)

print("Tabla Comparativa de KPIs por Tienda (Calificaciones sin redondear)")
display(styled_table)
```
